### Проблемы датасета:
    
* пропущенных значений явно нет, однако, судя по **event_id**, номерам индексов строк, а также датам подтоплений, *датасет неполный*: пропущенные подтопления между 2013 и 2018 годами и между 2018 и 2020 годами, возможно, пропущены и другие значения в рамках одного года (например, отсутствуют индексы 2, 3 за 2013 год)


* опечаток или дубликатов названий населенных пунктов, районов, субъекта нет


* странным кажется совпадение значений **house_count, farm_count** в первых двух строках: маловероятно, что в двух разных населенных пунктах в разные дни было подтоплено абсолютно одинаковое количество домов и ферм


* также вопросы вызывают строки, в которых совпадают значения **house_count** и **farm_count**. С одной стороны, возможно, были подтоплены дома с фермерскими участками, но, с другой стороны, возможно, произошло дублирование значений


* даты (**flood_date**) необходимо преобразовать в тип *datetime*, что позволить работать с датами более удобно. К тому же после преобразования они будут в стандартном формате ГГГГ-ММ-ДД


* переменные **house_count, farm_count** - типа *int*, а **bridge_count, road_count, social_object_count** - типа *float*. Имеет смысл преобразовать все эти переменные к одному типу *int*


* год является порядковой переменной, а не количественной, возможно, стоит рассматривать его как тип *object*, а не *int*. Однако если предполагаются какие-либо вычисления с датами, этот столбец можно преобразовать в *datetime*


* переменная **flood_type** имеет два уникальных значения, но на самом деле это одно значение - "паводок", написанное с большой буквы и с маленькой. Для дальнейшей работы с датасетом необходимо привести их к единому виду. Также видно, что разница в регистре связана с годом события: тип "Паводок" с большой буквы используется только в 2013 году
